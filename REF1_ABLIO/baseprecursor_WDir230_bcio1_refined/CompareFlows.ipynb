```python
# Start the amrwind_frontend app 
baserun = amrwind.MyApp.init_nogui()
bcio1   = amrwind.MyApp.init_nogui()
```

## Plot the domain

```python
bcio1.loadAMRWindInput('IEA29REF1_precursor_bcio1_refined.inp');
```

```python
# Set some options about what to plot
bcio1.popup_storteddata['plotdomain']['plot_windnortharrows'] = True
bcio1.popup_storteddata['plotdomain']['plot_sampleprobes']    = [] #['hhplane1','hhplane2','xzplane1' ] 
bcio1.popup_storteddata['plotdomain']['plot_refineboxes']     = ['box1']
bcio1.popup_storteddata['plotdomain']['plot_turbines']        = []
```

```python
# Plot the domain
plt.rc('font', size=14)
fig, axs = plt.subplots(3, 1, figsize=(10,20), facecolor='w')

bcio1.popup_storteddata['plotdomain']['plot_chooseview']      = 'XY'
bcio1.plotDomain(ax=axs[0])
#axs[0].set_xlim([-500, 1500])
#axs[0].set_ylim([-1000,1000])

bcio1.popup_storteddata['plotdomain']['plot_chooseview']      = 'XZ'
bcio1.plotDomain(ax=axs[1])

bcio1.popup_storteddata['plotdomain']['plot_chooseview']      = 'YZ'
bcio1.plotDomain(ax=axs[2])
```

## Plot the flow fields

```python
# Load the netcdf files
baserun.Samplepostpro_loadnetcdffile('../baseprecursor_WDir230/post_processing/sampling30000.nc')
bcio1.Samplepostpro_loadnetcdffile('post_processing/sampling30000.nc')
```

```python
# Plot an image
levels=np.linspace(0,7,41)
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(10,5), gridspec_kw={'width_ratios': [1, 1, 0.05]}, dpi=150)
im1 = baserun.plotSamplePlane('p_hub', 'velocityx',   6, 3, 'X','Y',ax=ax1, colorbar=False, levels=levels, cmap=cm.jet)
im2 = bcio1.plotSamplePlane('p_hub',   'velocityx',   6, 3, 'X','Y',ax=ax2, colorbar=False, levels=levels, cmap=cm.jet)
fig.colorbar(im1[0], cax=ax3)
fig.suptitle('Small IEA29 REF1 ABL, t=15300s, z=57m, Vx')
ax1.set_title('Base precursor run')
ax2.set_title('Using bndry_io_mode = 1')
ax2.get_yaxis().set_visible(False)
```

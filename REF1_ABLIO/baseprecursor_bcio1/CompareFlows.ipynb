```python
%%capture 

# Important header information
amrwindfedir = '/ascldap/users/lcheung/local/Git/amrwind-frontend'
import sys, os
sys.path.insert(1, amrwindfedir)

# Load the libraries
import matplotlib.pyplot    as plt
from matplotlib import cm
import amrwind_frontend as amrwind
import numpy as np

# Also ignore warnings
import warnings
warnings.filterwarnings('ignore')

# Make all plots inline 
%matplotlib inline

from matplotlib import animation
from IPython.display import HTML
```

```python
# Start the amrwind_frontend app 
baserun = amrwind.MyApp.init_nogui()
bcio1   = amrwind.MyApp.init_nogui()
```

```python
baserun.Samplepostpro_loadnetcdffile('../baseprecursor/post_processing/sampling00000.nc')
bcio1.Samplepostpro_loadnetcdffile('post_processing/sampling30000.nc')
```

```python
# Plot an image
levels=np.linspace(4,8,21)
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(10,5), gridspec_kw={'width_ratios': [1, 1, 0.05]}, dpi=150)
im1 = baserun.plotSamplePlane('p_hub', 'velocityx', 306, 3, 'X','Y',ax=ax1, colorbar=False, levels=levels, cmap=cm.jet)
im2 = bcio1.plotSamplePlane('p_hub',   'velocityx',   6, 3, 'X','Y',ax=ax2, colorbar=False, levels=levels, cmap=cm.jet)
fig.colorbar(im1[0], cax=ax3)
fig.suptitle('Small IEA29 REF1 ABL, t=15300s, z=57m, Vx')
ax1.set_title('Base precursor run')
ax2.set_title('Using bndry_io_mode = 1')
ax2.get_yaxis().set_visible(False)
```
